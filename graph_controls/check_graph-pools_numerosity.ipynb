Python notebook used to make sure that each folder for running the experiment has the correct number of graphs for each K value. This is to catch any error due to untracked files while creating or copying the graph pools across folders.

```python
import os

# Desktop directory where the folders are located
desktop_dir = "..\.."
N_values = ["100", "150", "200", "300", "400", "480", "600", "800", "1000"]
for N in N_values:
    # define the directory to check:
    directory = os.path.join(desktop_dir, "HUPLACLIP-" + N)
    # define the path to the folder to check:
    folder_path_clique = os.path.join(directory, "graph_pools", "sqrt_grid", "p_reduce", "N" + N.zfill(4), "CLIQUE")
    folder_path_noclique = os.path.join(directory, "graph_pools", "sqrt_grid", "p_reduce", "N" + N.zfill(4), "NO_CLIQUE")
    for folder_path in [folder_path_clique, folder_path_noclique]:
        # Check if the folder exists
        if os.path.exists(folder_path):
            # Get the list of files in each subfolder:
            subfolders = [f.path for f in os.scandir(folder_path) if f.is_dir()]
            for subfolder in subfolders:
                # Get the list of files in the subfolder
                file_list = os.listdir(subfolder)
                
                # Check if the number of files is exactly 100
                if len(file_list) != 100:
                    print(f"Folder '{subfolder}' has {len(file_list)} files (expected 100).")
                else:
                    print(f"Folder '{subfolder}' is correct.")
        else:
            print(f"Folder '{folder_path}' does not exist.")
            continue
```
